```python
import awswrangler as wr
import boto3
```

```python
def list_bucket_contents(bucket_resource, match='', size_mb=0):
    total_size_gb = 0
    total_files = 0
    match_size_gb = 0
    match_files = 0
    objects = []
    for key in bucket_resource.objects.all():
        key_size_mb = key.size/1024/1024
        total_size_gb += key_size_mb
        total_files += 1
        list_check = False
        if not match:
            list_check = True
        elif match in key.key:
            list_check = True

        if list_check and not size_mb:
            match_files += 1
            match_size_gb += key_size_mb
            
            print(f'{key.key} ({key_size_mb:3.0f}MB)')
            objects.append(key)            
            
        elif list_check and key_size_mb >= size_mb:
            match_files += 1
            match_size_gb += key_size_mb
            objects.append(key)
            print(f'{key.key} ({key_size_mb:3.0f}MB)')

    if match:
        print(f'Matched file size is {match_size_gb/1024:3.1f}GB with {match_files} files')            
    
    print(f'Bucket {bucket_resource.name} total size is {total_size_gb/1024:3.1f}GB with {total_files} files')
    return objects
```

```python
s3 = boto3.client('s3')
s3_resource = boto3.resource('s3')
taxi_bucket = s3_resource.Bucket("nyc-tlc")

# select all files with matching having "trip data" in S3 key and bigger than 1MB
objects = list_bucket_contents(taxi_bucket, match="green", size_mb=100)
```

```python
# let's pick random index
idx = 1
s3_data_obj = f"s3://{objects[idx].bucket_name}/{objects[idx].key}"
print(s3_data_obj)
```

```python
df = wr.s3.read_csv(s3_data_obj)
```

```python
#Let's drop columns which are doesn't have meaningful information

df = df.drop(["Trip_type ", "Payment_type", "Tip_amount"], axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (6, 4)
%matplotlib inline

color = sns.color_palette()
```

```python
fig = plt.figure(figsize=(28,14))
ax1 = fig.add_subplot(121)
ax1.scatter(df.Pickup_longitude,
            df.Pickup_latitude,
            s=1,alpha=0.1,color='red')

plt.ylim([40.60,41.00])
plt.xlim([-74.15,-73.70])
plt.xlabel('Longitude',fontsize=16)
plt.ylabel('Latitude',fontsize=16)
plt.title('Pickup Location',fontsize=18)
```

```python
numtrain = df.select_dtypes(include=[np.number])
corr = numtrain.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr,vmax=1,square=True,annot=True)
```

```python
df['Lpep_dropoff_datetime'] = pd.to_datetime(df['Lpep_dropoff_datetime'])
df['dropoff_hour'] = df.Lpep_dropoff_datetime.dt.hour
df['dropoff_week'] = df.Lpep_dropoff_datetime.dt.weekday
df['dropoff_month'] = df.Lpep_dropoff_datetime.dt.month
df['dropoff_day'] = df.Lpep_dropoff_datetime.dt.day
```

```python
df['dropoff_hour'] = df['dropoff_hour'].apply(int)
df['dropoff_week'] = df['dropoff_week'].apply(int)
df['dropoff_month'] = df['dropoff_month'].apply(int)
df['dropoff_day'] = df['dropoff_day'].apply(int)
```

```python
plt.figure(figsize=(8,6))
sns.pointplot(x='dropoff_hour',y='Trip_distance',data=df,kind='point',hue='pickup_month')
plt.xlabel('pickup_hour',fontsize=16)
plt.ylabel('mean(miles)',fontsize=16)
```
